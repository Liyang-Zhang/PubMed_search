```python
import os
import pandas as pd
import numpy as np

os.chdir('/dssg/home/acct-medkwf/medkwf4/script/yiyueUpdate')
dir_path = '/dssg/home/acct-medkwf/medkwf4/data/yiyue_Update/yiyue_WES/anno'
```

```python
# import candidate gene list
candidate_df = pd.read_excel('/dssg/home/acct-medkwf/medkwf4/script/yiyueUpdate/candidatesMerge.xlsx')
candidate_list = candidate_df['Gene'].tolist()
```

```python
# get a list of all Excel files in the directory
file_list = [f for f in os.listdir(dir_path) if f.endswith('.xlsx')]
```

```python
# store all the picked rows
merge_df = pd.DataFrame()

# loop over the list of files and read each one into a DataFrame
for file in file_list:
    file_path = os.path.join(dir_path, file)
    report_df = pd.read_excel(file_path)
    
    # 提取样本相关信息
    sampleName = file.split(".")[0]
    libraryName = file.split(".")[0].split("-")[1]
    reportName = file.split(".")[0].split("-")[1].replace("Lib", "A")
    print("file name: %s\nreport name: %s" % (file, reportName))
    
    # 寻找在list中的基因
    picked_df = report_df[report_df['Gene.refGene'].isin(candidate_list)]
    
    # 合并pubmed ID
    merge_single_df = pd.merge(picked_df, candidate_df, left_on='Gene.refGene', right_on='Gene', how='left')
    
    # Gene.refGene与Gene重复，删除Gene
    #merge_single_df = merge_single_df.drop('Gene', axis=1)
    
    # 加入sampleName，libraryName，reportName
    merge_single_df['sampleName'] = sampleName
    merge_single_df['libraryName'] = libraryName
    merge_single_df['reportName'] = reportName
    
    # Merge to the final dataframe
    merge_df = pd.concat([merge_df, merge_single_df])
```

```python
merge_df.to_excel('yiyueWES_pickedGenes_480sample.xlsx')
```
